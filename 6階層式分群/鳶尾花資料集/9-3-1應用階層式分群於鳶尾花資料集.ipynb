## 匯入資料

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import load_iris
iris=load_iris()
X=pd.DataFrame(iris.data)
X=X[[0,1]]
print(X.head())
```

## 檢查資料

```python
print(X.isnull().values.sum())
print("資料筆數:", X.shape)
print("資料的欄位名稱，分別是:", X.keys())
print("第一筆的資料內容:", X.iloc[0,::])
```

## 建立模型

```python
model=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
```

## 訓練與預測模型

```python
model.fit_predict(X)
```

## 繪製分群結果圖

```python
y = model.fit_predict(X)
plt.scatter(X[[0]],X[[1]],c= y)
```

## 更換模型計算距離公式

```python
model=AgglomerativeClustering(n_clusters=3,affinity='manhattan',linkage='average')
y=model.fit_predict(X)
plt.scatter(X[[0]],X[[1]],c=y)
```

## 使用scipy繪製階層式分群的樹狀圖

```python
import scipy.cluster.hierarchy as hier
model = hier.linkage(X,metric='euclidean',method='ward')
plt.figure(figsize=(12,8))
hier.dendrogram(model)
plt.title('Hierarchical Clustering')
plt.show()
```
